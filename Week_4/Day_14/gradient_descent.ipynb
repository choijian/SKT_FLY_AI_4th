```python
import torch
from torch.autograd import Variable

x = Variable(
 torch.tensor(1., dtype=torch.float32),
 requires_grad=True)
y = Variable(
 torch.tensor(1., dtype=torch.float32),
 requires_grad=True)
z = Variable(
 torch.tensor(1., dtype=torch.float32),
 requires_grad=True)
```

```python
x, y
```

```python
import math

optimizer = torch.optim.SGD(params=[x, y], lr=0.01)

EPOCHS = 100
for epoch in range(EPOCHS):
    f = 2*x**2 + y**2 + math.exp(4*x*y)
    optimizer.zero_grad()
    f.backward()
    optimizer.step()
```
