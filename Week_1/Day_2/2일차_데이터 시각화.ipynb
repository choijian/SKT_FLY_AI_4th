```python
x = np.arange(0,10)
y = x**2

plt.rc('font', family='NanumBarunGothic')
plt.plot(x,y)
plt.title('플롯 그래프')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.sin(x+0.5)
y3 = np.sin(x+1)

plt.rc('font', family='NanumBarunGothic')
plt.plot(x,y1,'r--')
plt.plot(x,y2,'b.')
plt.plot(x,y3, 'g-.')
plt.title('사인 함수')
plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.sin(x+0.5)
y3 = np.sin(x+1)

plt.rc('font', family='NanumBarunGothic')
plt.plot(x,y1,'r--', label='Sin A')
plt.plot(x,y2,'b.', label='Sin B')
plt.plot(x,y3, 'g-.', label = 'Sin C')
plt.legend(loc = 'upper right')
plt.title('사인 함수')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

plt.rc('font', family='NanumBarunGothic')
plt.style.use('ggplot')
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.plot(x,y,'r-')

plt.subplot(2,2,2)
plt.plot(x,z,'g--')

plt.subplot(2,2,3)
plt.plot(x,z,'b-.')

plt.show()
```

## 산점도

```python
x = np.random.rand(30)
y = np.random.rand(30)

plt.scatter(x,y)
plt.show()
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n)) ** 2
colors = np.random.rand(n)
plt.scatter(x,y, s=area, c=colors, alpha=0.7) #alpha=투명도
plt.show()
```

## 막대 그래프

```python
x = ['A','B','C','D']
y = [5. , 25. , 50. , 20.]
plt.bar(x, y)

plt.title('막대 그래프')
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

```python
name = ['A','B','C','D']
y = [5. , 25. , 50. , 20.]
z = [8. , 20. , 25. , 6.]
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(12, 5)
ax[0].bar(name, y, color='#33aaff')
ax[1].bar(name, z, color=['r', 'g', 'b', 'k'])
plt.show()
```

```python
x = np.random.normal(172,10,250)
plt.hist(x)
plt.show()
```

```python
x = np.random.normal(172,10,250)
plt.hist(x, bins=20, color='#05ae1c')
plt.show()
```

```python
x = np.random.normal(10, 3, 100)
y = np.random.normal(13, 3, 100)
z = np.random.normal(8, 5, 100)
plt.boxplot([x, y, z])
plt.xticks([1, 2, 3], ['A', 'B', 'C'])
plt.show()
```

```python
x = np.random.normal(10, 3, 100)
y = np.random.normal(13, 3, 100)
z = np.random.normal(8, 5, 100)
plt.boxplot([x, y, z], patch_artist=True, sym='ro')
plt.xticks([1, 2, 3], ['A', 'B', 'C'])
plt.show()

plt.savefig('box.png')
```
