# 학교 위치 찍기

```python
import pandas as pd
df = pd.read_excel('seoulUniv.xlsx')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
for name, lat, lng in zip(df.index, df.위도, df.경도):
  folium.CircleMarker([lat,lng], radius=10, color='brwon', fill=True, fill_color='coral', fill_opacity=0.7, popup=name).add_to(seoul_map)

seoul_map
```
