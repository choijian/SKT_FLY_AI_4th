```python
import copy #깊은 복사를 위한 library
```

```python
a = [1,2,3]
b = copy.deepcopy(a)
b[0] = 100
print(a,b)
```

```python
a = [1,2,3]
b = a[:] # 값만 복사
b[0] = 100
print(a,b)
```
